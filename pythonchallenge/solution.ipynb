```python
def print_count(num_s):
    counter = 0
    prev = None
    res = []
    for curr in num_s:
        if prev is None:
            prev = curr
            counter = 1
        elif prev == curr:
            counter+=1
        else:
            res.append(str(counter))
            res.append(prev)
            prev = curr
            counter = 1
    res.append(str(counter))
    res.append(prev)
    return res
```

```python
val = ['1']
for i in range(32):
    val = print_count(val)
    print('a[{}]'.format(i),  len(val))
```
